```python
import pandas as pd
```

```python
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

# Merge datasets
data = deliveries.merge(matches[['id','winner']], left_on='match_id', right_on='id')
```

```python
data
```

```python
data['cumulative_runs'] = data.groupby('match_id')['total_runs'].cumsum()
```

```python
data['cumulative_wickets'] = data.groupby('match_id')['is_wicket'].cumsum()
```

```python
data['runs_left'] = data['total_runs'] - data['cumulative_runs']
data['balls_left'] = 120 - (data['over']*6 + data['ball'])
data['wickets_left'] = 10 - data['cumulative_wickets']
```

```python
data.drop(columns=['extras_type','player_dismissed','dismissal_kind','fielder'],inplace=True)
```

```python
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['batting_team'] = le.fit_transform(data['batting_team'])
data['bowling_team'] = le.fit_transform(data['bowling_team'])
```

```python
data.drop(columns=['non_striker','bowler','batter'],inplace=True)
```

```python
data['winner'].value_counts()
```

```python
data['winner']=le.fit(data['winner']).transform(data['winner'])
```

```python
data['winner'].value_counts()
```

```python
data
```

```python
data.drop(columns=['id','crr','rrr','match_id'],inplace=True)
```

```python
data
```

```python
data['result'] = (data['batting_team'] == data['winner']).astype(int)
```

```python
data
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = data[['inning', 'batting_team', 'bowling_team', 'over', 'ball',
       'batsman_runs', 'extra_runs', 'total_runs', 'is_wicket',
       'cumulative_runs', 'cumulative_wickets', 'runs_left', 'balls_left',
       'wickets_left']]
y = data['result']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier()
}

# training + evaluation
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print(f"{name}: {acc:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# features and target
X = data[['inning', 'batting_team', 'bowling_team', 'over', 'ball',
       'batsman_runs', 'extra_runs', 'total_runs', 'is_wicket',
       'cumulative_runs', 'cumulative_wickets', 'runs_left', 'balls_left',
       'wickets_left']]
y = data['result']

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# model
dt = DecisionTreeClassifier()

# train
dt.fit(X_train, y_train)

# predict
y_pred_dt = dt.predict(X_test)

# accuracy
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# features and target
X = data[['inning', 'batting_team', 'bowling_team', 'over', 'ball',
       'batsman_runs', 'extra_runs', 'total_runs', 'is_wicket',
       'cumulative_runs', 'cumulative_wickets', 'runs_left', 'balls_left',
       'wickets_left']]
y = data['result']

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# model
rf = RandomForestClassifier()

# train
rf.fit(X_train, y_train)

# predict
y_pred_rf = rf.predict(X_test)

# accuracy
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```
